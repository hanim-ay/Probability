```python
########################################
#Faktoriyel
########################################
def faktoriyel(sayi):
    fact = 1 
    if sayi < 0: 
        print ("Ltfen pozitif bir say giriniz.")
    else:
        for i in range(1,sayi + 1): 
            fact = fact*i 
    return fact
```

```python
Istenilen_durumlar = faktoriyel(3)*faktoriyel(6)
Tum_durumlar = faktoriyel(8)
Olaslk = Istenilen_durumlar/Tum_durumlar
print(Olaslk)
```
