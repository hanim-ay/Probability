**Permutasyon ve Kombinasyon Örnekleri**


Kullanılan fonksiyonlar aşağıda gösterilmiştir.

```python
########################################
#Faktoriyel
########################################
def faktoriyel(sayi):
    fact = 1 
    if sayi < 0: 
        print ("Lütfen pozitif bir sayı giriniz.")
    else:
        for i in range(1,sayi + 1): 
            fact = fact*i 
    return fact
########################################
#Permutasyon
########################################
def permutasyon(sayi1,sayi2):
    perm = 0
    if (sayi2>sayi1):
         print ("Lütfen eleman sayısından (n) küçük bir seçim sayısı (r) giriniz.")
    else:
        perm = faktoriyel(sayi1)/faktoriyel(sayi1-sayi2)
    return perm
########################################
#Kombinasyon
########################################
def kombinasyon(sayi1,sayi2):
    comb = 0
    if (sayi2>sayi1):
        print ("Lütfen eleman sayısından (n) küçük bir seçim sayısı (r) giriniz.")
    else:
        comb = faktoriyel(sayi1)/(faktoriyel(sayi1-sayi2)*faktoriyel(sayi2))
    return comb
```
